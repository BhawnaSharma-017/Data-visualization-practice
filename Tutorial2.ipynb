```python
data = {
    'Salary' :[23000,34787,33245,21233,34900,23345,67889,23324,
               34466,56679,98700,32450,45654,77890,34546,67800,
               34500,34560,76700,34300]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.shape
```

```python
## Lineplot
plt.plot(df['Salary'],color = 'Red',marker = 'o',linestyle = ':',
         linewidth = 2)
plt.grid(True)
plt.show()
```

```python
# Histogram
plt.hist(df['Salary'],bins = 5,color = 'Green')
plt.show()
```

```python
# Boxplot
plt.boxplot(df['Salary'])
plt.show()
```

```python
# Univariate : Categorical Column
```

```python
df['dept'] = ['HR','IT','Finance','HR','Finance','IT','HR',
              'Finance','IT','HR'] * 2
```

```python
df.head()
```

```python
df.info()
```

```python
count = df['dept'].value_counts()
count
```

```python
## Pie chart
plt.pie(count,labels = count.index,autopct = "%1.1f",
        explode = [0,0.1,0])
plt.show
```

```python
# Countplot
plt.bar(count.index,count , color = ['Green','Red','Black'])
plt.show()
```

##### Bivariate -> Numerical-Categorical

```python
hr_sal = df[df['dept'] == 'HR']['Salary']
```

```python
hr_sal
```

```python
finance_sal = df[df['dept'] == 'Finance']['Salary']
finance_sal
```

```python
IT_sal = df[df['dept'] == 'IT']['Salary']
IT_sal
```

```python
## Boxplot

plt.boxplot([finance_sal,IT_sal,hr_sal] , labels = ['Finance',
                                                    'IT','HR'])
plt.show()
```

```python
salary_by_dept = df.groupby('dept')['Salary'].sum()
```

```python
salary_by_dept
```

```python
## Pie chart
plt.pie(salary_by_dept,labels = salary_by_dept.index,
        autopct = "%1.2f")
plt.axis('equal')
plt.show()
```

```python
## Bar plot
hr_mean = sum(hr_sal) / len(hr_sal)
fin_mean = sum(finance_sal) / len(finance_sal)
it_mean = sum(IT_sal) / len(IT_sal)
```

```python
plt.bar(['HR','IT','Finance'],[hr_mean,it_mean,fin_mean]
        ,color = ['Green','Yellow','Purple'])
plt.show()
```

#### Multivariate Analysis :-> Three numerical column

```python
df['experience'] = [1.3,2,4,5,6,2.2,3.4,7,7.5,8,
                    8.2,9,5,1,4,4.3,3.6,5,5.6,9]
df.head()
```

```python
plt.pie(salary_by_dept,labels = salary_by_dept.index,
        autopct = '%1.2f',shadow = True,explode=[0.1,0,0.1])
plt.axis('equal')
plt.show()
```

```python
## Bubble plot
plt.scatter(df['Age'],df['Salary'],df['experience']*10,
            color = 'violet',edgecolors='purple')
plt.show()
```

```python
## 2 numerical and 1 categorical column
plt.scatter(df['Age'],df['Salary'],
            c = df['dept'].map({'HR':'yellow','IT':'Blue',
                                'Finance':'Orange'}),
            cmap = 'rainbow')
plt.show()
```

```python
data = {
    'Year' :[2020,2021,2022,2023],
    'Sales' : [100,150,200,250],
    'Profit' : [20,30,40,50],
    'Expenses' : [80,120,160,200]
}
df2 = pd.DataFrame(data)
```

```python
df2.head()
```

```python
plt.plot(df2['Year'],df2['Sales'],label = 'Sales')
plt.plot(df2['Year'],df2['Profit'],label = 'Profit')
plt.plot(df2['Year'],df2['Expenses'],label = 'Expenses')
plt.title("Financial Analysis")
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
## 3-D Plot
ax = plt.axes(projection = '3d')

ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_zlabel('Experience')
ax.scatter(df['Age'],df['Salary'],df['experience'])
plt.show()
```
