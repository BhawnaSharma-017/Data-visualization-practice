```python
import seaborn as sms
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
sms.get_dataset_names()
```

```python
penguins = sms.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
sms.set_style('darkgrid')  # whitegrid , ticks , dark , white , darkgrid  
sms.scatterplot(data = penguins , x='flipper_length_mm', 
                y='body_mass_g' ,hue = 'sex')
sms.despine()
```

```python
sms.scatterplot(data = penguins , x='flipper_length_mm',
                 y='body_mass_g' ,hue = 'island' , palette="Set1")  # palette = Set1,Set2,Set3
sms.set_context("paper")  # paper , notebook , talk , poster
```

```python
sms.scatterplot(data = penguins,x = "species" , y= "body_mass_g" 
                ,hue = "island" ,style = "sex")
```

```python
# Strip plot
sms.stripplot(data = penguins,x = "species" , y = "body_mass_g" 
              , hue = "island" ,dodge=True)  
# dodge = it will seperate my data-points according to my hue
```

```python
#swarm plot
sms.set_context("talk")
sms.swarmplot(data = penguins , x = 'species' , y = "body_mass_g" 
              , hue = "island")
sms.despine()
```

```python
# histogram
sms.histplot(data = penguins , x = 'body_mass_g' , hue = "sex"
            ,palette='Set1')
```

```python
# Regression plot :
sms.regplot(data = penguins , x = "body_mass_g" , 
            y = "flipper_length_mm")
```

```python
sms.regplot(data = penguins , x = "body_mass_g" ,
             y = "flipper_length_mm" , scatter=False)
```

```python
# line plot
sms.set_style("whitegrid")
sms.lineplot(data = penguins , x = "body_mass_g" , 
             y = "flipper_length_mm" , hue = "island",
             style = "sex",palette='Set2')
```

```python
# joinplot -- To combine two different plots
sms.jointplot(data=penguins , x = "body_mass_g" , y = "flipper_length_mm")
```

```python
sms.jointplot(data=penguins , x = "body_mass_g" , y = "flipper_length_mm" , hue = "sex")
```

```python
sms.jointplot(data=penguins , x = "body_mass_g" , y = "flipper_length_mm" , 
              hue = "sex" , kind = "kde")
```

```python
sms.jointplot(data=penguins , x = "body_mass_g" , y = "flipper_length_mm" ,kind = "reg")
```

```python
# Bar plot
# estimator = np.mean-- By default :- more :-> (sum,median,std,var)
sms.barplot(data=penguins , x = "species" , y = "body_mass_g" , hue = "sex" , 
            palette=['green','skyblue'], estimator = np.std)
```

```python
# countplot 
sms.countplot(data = penguins , x = 'species' ,hue = "island" , palette="Set2")
```

```python
# box plot  - lower extreme , lower quartile , median , upper quartile , upper extreme
sms.boxplot(data = penguins , x = 'species' , y = "body_mass_g" , hue="sex")
```

```python
# violin plot - min_value , max_value , mid_value
sms.violinplot(data =  penguins,x = "species" , y="body_mass_g",hue="sex" )
```

```python
# Split method in violinplot
# Width of the violin : Density of the data at different values
# Middle line : Median
# Thick Bar : Intermediate range
sms.violinplot(data =  penguins,x = "species" , y="body_mass_g",hue="sex",split = True)
```

```python
sms.violinplot(data =  penguins,x = "species" , y="body_mass_g",palette="pastel" )
sms.swarmplot(data =  penguins,x = "species" , y="body_mass_g" , color = "black" ,
               size = 3 , hue ="island" ,palette="dark")
```

```python
# KDE (Kernal Density Estiamtion)-> for estimating th probability density function of 
# continuous or non-perimetric data
# Or in simple terms : a smooth curve that shows the distribution of data 
# - like a smoothed-out version
# It helps you see where the data is concentrated (peaks) and how it spreads  
# (like a histogram)
sms.kdeplot(data =  penguins,x="body_mass_g",hue = "species",fill = True) 
# fill = False not fill the graph curve colors
```

```python
penguins.head()
```

```python
# corr() - correation function
columns  = ["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
penguins[columns].corr()
```

```python
# heatmap
sms.heatmap(data  =penguins[columns].corr(),annot = True,cmap = "Blues",linewidth  ="2") 
# annot =  True - gives an numerical value on each box (labeled the boxes)
```

```python
# rugplot  - to give extra details

sms.rugplot(data  = penguins,x = "body_mass_g",hue  ="species",palette="Set2")
```

```python

# for y axis
sms.rugplot(data  = penguins,y = "body_mass_g",hue  ="species",palette="Set2")
```

```python
sms.rugplot(data  = penguins,y = "body_mass_g",hue  ="species",palette="Set2" ,
             height  =0.2)
```

```python
# pair plot - to show the relationship in a grid of datasets  (here a diagonal and a lower-triangle and upper-triangle)
# To visulaize the all numerical variable together (by default take numerical data)
sms.pairplot(data = penguins,hue  ="sex",palette="Set2",diag_kind="kde") 
 # diag means diagonal line graph has "kde" kind of graphs
```

```python
# pair grid -> subplot grid for plotting pairwise relationships in a dataset. 
graph = sms.PairGrid(data = penguins,hue = "sex",palette="Set2")
graph.map_upper(sms.scatterplot)
graph.map_lower(sms.kdeplot)
graph.map_diag(sms.histplot)
graph.add_legend()
```
