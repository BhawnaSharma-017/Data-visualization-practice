```python
month = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(month,sales,color = 'blue',linewidth = 2 ,
         linestyle = '--', marker = 'o', 
         label = '2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title("Monthly Sales Data Report")
plt.legend(loc = "upper left",fontsize = 12)
plt.grid(color = 'grey' , linestyle = ':' , linewidth = 1)
plt.show()
```

Bar Chart

plt.bar(x , y , height , color , width , label)

```python
product = ['A','B','C','D']
sales = [1000,1500,800,1200]
```

```python
plt.bar(product,sales,color='orange',label = 'Sales 2025')
plt.xlabel('Product')
plt.ylabel("Sales")
plt.title('Product sale Comparsion')
plt.legend()
plt.show()
```

```python
plt.barh(product,sales,color='orange',label = 'Sales 2025')
plt.xlabel('Product')
plt.ylabel("Sales")
plt.title('Product sale Comparsion')
plt.legend()
plt.show()
```

```python
regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]
# %1.1f%% means format string contain one digit before decimal, 
# one digit after decimal
plt.pie(revenue,labels = regions ,autopct='%1.1f%%', 
        colors = ['gold','skyblue','lightgreen','coral'])
plt.title('Revenue Contribution by region')
plt.show()
```

### Histogram  
For numerical continuous data , not for categorical data

```python
scores = [45,67,89,56,78,88,92,60,74,81,55,66,75,82,90,85,70,73,68,77]
plt.hist(scores , bins=5,color = 'purple',edgecolor = 'black')
plt.xlabel('Score range')
plt.ylabel('Number of students')
plt.title('Score Distributions of Students')
plt.show()
```

### Scatter Plot 
- A graph where each decimal point represent each observation with each value
- To find the relationship between two data list

```python
# plt.scatter(x , y, color ,marker , label )
import matplotlib.pyplot as plt
```

```python
hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

plt.scatter(hours_studies,exam_scores,color = 'green',
            marker = 'o',label='Student data')
plt.xlabel('Hours studied')
plt.ylabel('Exam Score')
plt.title("Relationship b/w Study Time & Exam Score")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.scatter([1,2,3],[50,60,70],color = 'blue',marker = 'x',
            label='Student data A')
plt.scatter([1,2,3],[45,50,52],color = 'red',marker = 'D',
            label='Student data B')
plt.xlabel('Hours studied')
plt.ylabel('Exam Score')
plt.title("Comparasion of two Classes")
plt.grid(True)
plt.legend()
plt.show()
```

Subplot :- Two or more graphs in one plot
1. Multiple Charts
2. Space (less)
3. Quickly analyze data

```python
# plt.subplot(nrow,ncols,index)
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]
```

```python
plt.subplot(1,2,1)   #(1st row,2nd column,Ist position)
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)  #(1st row,2nd column,2nd position)
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python
# fig , ax = plt.subplots(nrows,ncols,figsize=(width,height)
fig, ax =plt.subplots(1,2,figsize=(10,5)) #(n-rows,n-cols,figsize)

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y,color = 'Green')
ax[0].set_title("Line plot")

ax[1].bar(x,y ,color='Red')
ax[1].set_title('Bar chart')

fig.suptitle("Comparasion  of Line & Bar Chart")
plt.tight_layout()
plt.show()
```

- plt.tight_layout()  :- To fix overlap
- savefig() :- To save our graph figure in any file e.g. :- .pdf, .png, .jpg
