```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv("top_rated_movies.csv")
df
```

```python
df = df.dropna(subset=['id','title','overview','release_date',
                       'popularity','vote_average','vote_count'])
```

```python
type_counts = df['title'].value_counts()
```

```python
plt.figure(figsize = (6,4))
plt.bar(type_counts.index,type_counts.values,color = ['skyblue','orange'])
plt.title("Number of Movies Vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig('movies_vs_tvshows.png')
plt.show()
```

```python
rating_counts = df['vote_count'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_counts,labels = rating_counts.index , 
        autopct='%1.1f%%',startangle=90)
plt.title("Percentage of content ratings")
plt.tight_layout()
plt.savefig('content_rating.png')
plt.show()
```

```python
release_counts = df['release_date'].value_counts().sort_index()
```

```python
plt.figure(figsize = (10,6))
plt.scatter(release_counts.index,release_counts.values,
            color='red')
plt.title("Release year VS Number of Shows")
plt.xlabel("Release year")
plt.ylabel("Number of Shows")
plt.tight_layout()
plt.savefig("movie_duration.png")
plt.show()
```
