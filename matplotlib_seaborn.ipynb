#### How can we compare the top 3 platforms the publisher use ?

```python
df.head()
```

#### Categorical data

#### Top 3 Publisher 

```python
top_3_pub = df['Publisher'].value_counts().index[:3]
top_3_pub
```

#### Top 3 Genre

```python
top_3_genre = df['Genre'].value_counts().index[:3]
top_3_genre
```

#### Top 3 Platforms

```python
top_3_plat = df['Platform'].value_counts().index[:3]
top_3_plat
```

```python
top_3_data = df.loc[(df['Publisher'].isin(top_3_pub)
                    &(df['Platform'].isin(top_3_plat))
                    &(df['Genre'].isin(top_3_genre)))]
```

```python
top_3_data
```

```python
sns.countplot(data = top_3_data,x='Publisher',hue = 'Platform')
# plt.xticks(rotation = 90)
```

#### Dogde plot - Stacked Count plot

```python
df_stacked_plt = pd.crosstab(index=top_3_data['Publisher'],
                             columns = top_3_data['Platform'])
```

```python
df_stacked_plt
```

```python
df_stacked_plt.plot(kind = 'bar' , stacked = True)
plt.show()
```

- When we want to study about statistical analysis of each category in the data
- Box plot can be used there

```python
sns.boxplot(x = 'Publisher',y= 'Global_Sales',data = top_3_data)
plt.show()
```
